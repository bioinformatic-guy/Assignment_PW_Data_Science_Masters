## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible Python web framework that allows developers to build web applications quickly and easily. It is a micro-framework, which means that it provides only the basic tools and features needed to build a web application, without imposing any particular way of doing things. Flask is designed to be easy to use, easy to learn, and highly extensible.

Some of the advantages of using Flask include:

1. Simplicity: Flask is easy to learn and use, making it a good choice for small to medium-sized web applications.

2. Flexibility: Flask allows developers to choose the tools and libraries that work best for their application, without being constrained by the framework.

3. Extensibility: Flask has a rich ecosystem of extensions that can be used to add new features to an application.

4. Minimalistic: Flask is a lightweight framework that has a minimalistic codebase, making it easy to understand and customize.

5. Testing: Flask provides an excellent testing framework that allows developers to test their application easily.

6. Performance: Flask has a low overhead and high performance, making it a good choice for applications that need to handle a large number of requests.

7. Documentation: Flask has excellent documentation that is easy to read and understand, making it easy for developers to get started with the framework.

Overall, Flask is a great choice for building web applications that are lightweight, flexible, and easy to maintain. Its simplicity, flexibility, and extensibility make it a popular choice among developers.

## Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping a URL or a URL pattern to a view function that handles the request and generates a response. It is done using the `@app.route()` decorator, which specifies the URL or URL pattern for a view function.

We use app routes in Flask to define the endpoints of our web application. Each endpoint corresponds to a specific URL, and when a user visits that URL, the associated view function is called to generate a response. App routes allow us to create a clean and organized URL structure for our web application, which makes it easy to navigate and understand.

In addition, app routes also provide a way to separate the concerns of our web application. By defining separate routes for different parts of our application, we can keep the logic for each part organized and easy to maintain. This can make it easier to add new features and fix bugs in our application over time.

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the `url_for()` function is used for URL building. It generates a URL for the specified endpoint with the given method and arguments.

```python
pip install flask
```

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

@app.route('/user/<username>')
def profile(username):
    return f'{username}\'s profile'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('hello', name='Alice'))
        print(url_for('profile', username='Bob'))
```

In this example, we have defined two routes: `/hello/<name>` and `/user/<username>.` The `hello()` function takes a name argument and returns a personalized greeting, while the `profile()` function takes a username argument and returns the user's profile.

We use the `url_for()` function to generate URLs for these routes. The first argument to `url_for()` is the name of the endpoint we want to generate a URL for (hello or profile in this case). We can also pass any additional arguments that the endpoint requires (e.g. name or username).

When we run the application and visit the `/hello/Alice` and `/user/Bob` URLs, we should see the personalized greetings and profile as expected. In addition, the `print()` statements in the `with app.test_request_context()` block will print the generated URLs to the console
