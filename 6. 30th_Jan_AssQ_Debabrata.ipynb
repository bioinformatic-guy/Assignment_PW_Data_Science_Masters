## Q4. Write a python code to check how many time a given number can be divided by 3 before it is less than or equal to 10.

```python
def divide_by_3(num):
    count = 0
    while num > 10:
        num = num / 3
        count += 1
    return count
```

```python
divide_by_3(99)
```

## Q5. Why and When to use `while` loop in Python give a detailed description with example

The `while` loop in Python is used to execute a block of code repeatedly as long as a given condition is true. 
The condition is evaluated before each iteration of the loop, and if it is `True`, the code inside the loop is executed. If the condition is `False,` the loop stops and execution continues with the code following the loop.

Here's an example of using a `while` loop to count the number of times a user enters an even number:

```python
count = 0
while True:
    num = int(input("Enter an even number (enter -1 to exit): "))
    if num == -1:
        break
    elif num % 2 == 0:
        count += 1

print("You entered", count, "even numbers.")
```

In this example, the while loop is set to run indefinitely because the condition True is always True. The loop will only exit when the break statement is executed inside the loop, which happens when the user enters -1. If the user enters an even number, the count variable is incremented. Finally, after the loop has finished, the program prints the number of even numbers entered by the user.

The while loop is particularly useful when you don't know beforehand how many times you need to loop. For example, you can use a while loop to keep reading from a file until there is no more data to read.

## Q6. Use nested while loop to print 3 different pattern

```python
#Triangle Pattern
rows = 5
for i in range(1, rows + 1):
    for j in range(1, rows - i + 1):
        print(end=" ")
    for k in range(1, i * 2):
        print("*", end="")
    print("")
```

```python
# Reverse Triangle Pattern
rows = 5
for i in range(rows, 0, -1):
    for j in range(1, rows - i + 1):
        print(end=" ")
    for k in range(1, i * 2):
        print("*", end="")
    print("")
```

```python
#Diamond Pattern
rows = 5
for i in range(1, rows + 1):
    for j in range(1, rows - i + 1):
        print(end=" ")
    for k in range(1, i * 2):
        print("*", end="")
    print("")
for i in range(rows - 1, 0, -1):
    for j in range(1, rows - i + 1):
        print(end=" ")
    for k in range(1, i * 2):
        print("*", end="")
    print("")
```

## Q7. Reverse a while loop to display number from 10 to 1

```python
num = 10
while num >= 1:
    print(num)
    num -= 1
```
