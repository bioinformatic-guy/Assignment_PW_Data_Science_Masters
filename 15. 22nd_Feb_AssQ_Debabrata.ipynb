## Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In Pandas, a DataFrame is a two-dimensional tabular data structure that consists of rows and columns. It is a table-like data structure that stores data in a rectangular format, where each row represents an observation or record, and each column represents a variable or feature. It can be thought of as a collection of Pandas Series, where each Series represents a column of data.

On the other hand, a Pandas Series is a one-dimensional labeled array that can hold data of any type (integer, float, string, etc.). It can be thought of as a single column of data, where each element is associated with a label or index.

```python
# Create a Pandas Series
s = pd.Series([1, 3, 5, 7, 9])

# Create a Pandas DataFrame
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]})

# Print the Series and the DataFrame
print(s)
print(df)
```

## Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

Pandas DataFrame provides a wide range of functions that can be used to manipulate and transform data. Some common functions include:

1. head() and tail(): These functions are used to view the first or last few rows of a DataFrame, respectively. They are useful for quickly inspecting the data.

2. describe(): This function provides a summary of the descriptive statistics of the DataFrame, such as count, mean, standard deviation, minimum and maximum values, etc.

3. loc[] and iloc[]: These functions are used to subset the DataFrame based on row and column indexes, respectively. loc[] uses labels to select rows or columns, while iloc[] uses integer positions.

4. dropna(): This function is used to remove rows or columns with missing values from the DataFrame.

## Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

Both Series and DataFrame are mutable in nature in Pandas. The data in both Series and DataFrame can be modified after they are created. However, Panel is an immutable data structure in Pandas, which means that once a Panel is created, its data cannot be changed. However, it is important to note that the Panel data structure has been deprecated since Pandas version 0.25.0 and is not recommended for use. Instead, the Panel functionality is mostly replaced by the more powerful MultiIndex objects.

## Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# create a Pandas Series for each column
names = pd.Series(['Alice', 'Bob', 'Charlie', 'David'])
ages = pd.Series([25, 30, 35,40])
countries = pd.Series(['USA', 'Canada', 'Mexico', 'Brazil'])

# combine the Series into a DataFrame
df = pd.DataFrame({'Name': names, 'Age': ages, 'Country': countries})

# print the DataFrame
print(df)
```
