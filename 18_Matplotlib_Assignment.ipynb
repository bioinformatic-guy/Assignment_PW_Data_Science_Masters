## Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**Matplotlib** is a popular data visualization library in Python. It provides a wide range of customizable and high-quality 2D and 3D plots for visualizing data.

`Matplotlib` is used for data visualization in various scientific and engineering fields, as well as in business analytics, finance, and many other domains. It allows users to create publication-quality charts and graphs easily, and also provides a wide range of customization options to suit the specific needs of the user.

Some of the key features of Matplotlib include support for multiple data formats, a wide range of plot types and styles, interactive visualization, animation, and support for multiple operating systems and programming languages.

Five common plot types that can be plotted using the Pyplot module of Matplotlib are:


1. Line plot - Used to display the relationship between two variables over a continuous interval. It is created using the plot() function in Matplotlib.

2. Scatter plot - Used to display the relationship between two variables as a collection of points. It is created using the `scatter()` function in Matplotlib.

3. Bar plot - Used to compare the values of different categories. It is created using the `bar()` or `barh(`) function in Matplotlib.

4. Histogram - Used to show the distribution of a single variable. It is created using the *hist()* function in Matplotlib.

5. Box plot - Used to show the distribution of a dataset along with its outliers. It is created using the `boxplot()` function in Matplotlib.
  







## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is a type of data visualization that displays the relationship between two numerical variables. It consists of a set of points, each representing a combination of values for the two variables being plotted, and is used to show the strength and direction of the relationship between the variables.

```python
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Random Scatter Plot')
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
### Use the following data:

The `subplot()` function is used to create multiple plots within a single figure. It allows you to divide the figure into a grid of rows and columns, and then specify which grid cell each plot should occupy. This is useful when you want to compare multiple plots side-by-side or arrange them in a specific layout.

```python
import numpy as np

# Create a figure with 2 rows and 2 columns of plots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))


#For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
#For line 2: 
y2 = np.array([50, 20, 40, 20, 60, 70])
#For line 3: 
y3 = np.array([10, 20, 30, 40, 50, 60])
#For line 4: 
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plot the data on each subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Plot 1')
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Plot 2')
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Plot 3')
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Plot 4')

# Add some labels and a title
fig.suptitle('Four Line Plots')
fig.tight_layout()

# Show the plot
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of chart that represents categorical data with rectangular bars, with the height or length of each bar proportional to the value it represents. It is used to compare the values of different categories or groups in a dataset, and is particularly useful for displaying data that can be easily categorized into distinct groups.

A bar plot can be used to show the distribution of a single variable, or to compare the values of multiple variables within each category. It is a simple and effective way to display data and can be easily customized to meet specific needs.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Create a bar plot
plt.bar(company, profit, color='blue', alpha=0.5)
 
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Bar Plot')
```

```python
# Create a horizontal bar plot
plt.barh(company, profit, color='blue', alpha=0.5)
 
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Horizontal Bar Plot')
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2])
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.title('Box Plot')
```
